```python
# converting the tweet to lower case
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['text'].head()
```

```python
#Removing stopwords
df['text'] = df['text'].apply(lambda x : ' '.join([word for word in x.split() if not word in set(stopwords.words('english'))]))
```

```python
#Lemmatization
lemmatizer = WordNetLemmatizer()
df['text'] = df['text'].apply(lambda x : ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
```

```python
#Stemming
ps = PorterStemmer()
df['text'] = df['text'].apply(lambda x : ' '.join([ps.stem(word) for word in x.split()]))
```

```python
# spelling correction
from textblob import TextBlob
df['text'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

```python
# remove !, ,, . , numbers 
def clean_tweets(tweet):
    # remove special characters 
    tweet = re.sub('[^ a-zA-Z0-9]', '', tweet)
    # remove Numbers
    tweet = re.sub('[0-9]', '', tweet)
    return tweet
df['text'] = df['text'].apply(clean_tweets)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
def tokenize(text):
    return word_tokenize(text)
df['tokenized'] = df['text'].apply(tokenize)
df.head()
```

```python
# 10 most common words
words = Counter()
for idx in df.index:
    words.update(df.loc[idx, "tokenized"])
words.most_common(10)
```

```python
# label encoding
df['sentiment']=df.sentiment.map({'positive':1,'neutral':1,'negative':0})
```

### Modelling  

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier
```

```python
X1_train, X1_test, Y1_train, Y1_test = train_test_split(df['text'], df['sentiment'], train_size = 0.75, random_state=0, shuffle = True, stratify=df['sentiment'])
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X1_train)
X_train_counts.shape
```

```python
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape
```

### Count Vectorizer + tfidf + SGD 

```python
text_clf_svm = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()),
                         ('clf-svm', SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, max_iter=5, random_state=42))])

text_clf_svm = text_clf_svm.fit(X1_train, Y1_train)
y_pred = text_clf_svm.predict(X1_test)
print(pd.crosstab(Y1_test,y_pred,rownames=['Actual'],colnames=['Predicted']))
print(classification_report(Y1_test, y_pred))
print(accuracy_score(Y1_test,y_pred))
from sklearn.metrics import roc_curve, auc
fp, tp, thresholds = roc_curve(Y1_test, y_pred)
roc_auc = auc(fp, tp)
plt.title('Receiver Operating Characteristic')
plt.plot(fp, tp, 'b', label = 'AUC on original = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

### Count Vectorizer + tfidf + RandomForest 

```python
from sklearn.pipeline import Pipeline
text_clf = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('clf', RandomForestClassifier())])
text_clf = text_clf.fit(X1_train, Y1_train)
y_pred = text_clf.predict(X1_test)
print(pd.crosstab(Y1_test,y_pred,rownames=['Actual'],colnames=['Predicted']))
print(classification_report(Y1_test, y_pred))
print(accuracy_score(Y1_test,y_pred))
from sklearn.metrics import roc_curve, auc
fp, tp, thresholds = roc_curve(Y1_test, y_pred)
roc_auc = auc(fp, tp)
plt.title('Receiver Operating Characteristic')
plt.plot(fp, tp, 'b', label = 'AUC on original = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

### Deep Learning Models 

```python
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.layers import TimeDistributed
```

```python
from sklearn.model_selection import train_test_split
text = df['text'].values
labels = df['sentiment'].values
text_train, text_test, y_train, y_test = train_test_split(text, labels)
print(text_train.shape, text_test.shape, y_train.shape, y_test.shape)
```

```python
vocab_size = 20000
maxlen = 500
embedding_size = 32

tokenizer = Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(text)

X_train = tokenizer.texts_to_sequences(text_train)
x_test = tokenizer.texts_to_sequences(text_test)

X_train = pad_sequences(X_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)
```

### LSTM Model 

```python
model = Sequential()
model.add(Embedding(vocab_size, embedding_size, input_length=maxlen))
model.add(LSTM(128))
#model.add(Dense(64, activation='relu'))
#model.add(Dense(1, activation='softmax'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        min_delta=1e-4,
        patience=3,
        verbose=1
    ),
    ModelCheckpoint(
        filepath='weights.h5',
        monitor='val_accuracy', 
        mode='max', 
        save_best_only=True,
        save_weights_only=True,
        verbose=1
    )
]
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, batch_size=30, validation_data=(x_test, y_test), epochs=2, callbacks=callbacks)
```

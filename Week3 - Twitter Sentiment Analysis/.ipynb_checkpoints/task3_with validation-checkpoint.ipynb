## Data Cleaning 

```python
## check which column has the null value
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
# converting the tweet to lower case
df['text'] = df['text'].apply(lambda x: " ".join([x.lower() for x in x.split()]))
df1['text'] = df1['text'].apply(lambda x: " ".join([x.lower() for x in x.split()]))
df['text'].head()
```

```python
df['text'] = df['text'].apply(lambda x : ' '.join([word for word in x.split() if not word in set(stopwords.words('english'))]))
## for validation set
df1['text'] = df1['text'].apply(lambda x : ' '.join([word for word in x.split() if not word in set(stopwords.words('english'))]))
```

```python
#Lemmatization
lemmatizer = WordNetLemmatizer()
df['text'] = df['text'].apply(lambda x : ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
df1['text'] = df1['text'].apply(lambda x : ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
```

```python
#Stemming
ps = PorterStemmer()
df['text'] = df['text'].apply(lambda x : ' '.join([ps.stem(word) for word in x.split()]))
df1['text'] = df1['text'].apply(lambda x : ' '.join([ps.stem(word) for word in x.split()]))
```

```python
# spelling correction
from textblob import TextBlob
df['text'][:5].apply(lambda x: str(TextBlob(x).correct()))
df1['text'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

```python
# remove !, ,, . , numbers 
def clean_tweets(tweet):
    # remove special characters 
    tweet = re.sub('[^ a-zA-Z0-9]', '', tweet)
    # remove Numbers
    tweet = re.sub('[0-9]', '', tweet)
    tweet = re.sub(r'^https?:\/\/.*[\r\n]*', '', tweet, flags=re.MULTILINE)
    return tweet
df['text'] = df['text'].apply(clean_tweets)
df1['text'] = df1['text'].apply(clean_tweets)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
def tokenize(text):
    return word_tokenize(text)
df['tokenized'] = df['text'].apply(tokenize)
df.head()
```

```python
# 10 most common words
words = Counter()
for idx in df.index:
    words.update(df.loc[idx, "tokenized"])
words.most_common(10)
```

```python
normal_words = ' '.join([word for word in df['text'][df['sentiment'] == 'neutral']])
wordcloud = WordCloud(width = 800, height = 500, max_font_size = 110,max_words = 100).generate(normal_words)
print('Normal words')
plt.figure(figsize= (12,8))
plt.imshow(wordcloud, interpolation = 'bilinear',cmap='viridis')
plt.axis('off')
```

```python
positive_words = ' '.join([word for word in df['text'][df['sentiment'] == 'positive']])
wordcloud = WordCloud(width = 800, height = 500, max_font_size = 110,max_words = 100).generate(normal_words)
print('Positive words')
plt.figure(figsize= (12,8))
plt.imshow(wordcloud, interpolation = 'bilinear',cmap='viridis')
plt.axis('off')
```

```python
negative_words = ' '.join([word for word in df['text'][df['sentiment'] == 'negative']])
wordcloud = WordCloud(width = 800, height = 500, max_font_size = 110,max_words = 100).generate(normal_words)
print('Negative words')
plt.figure(figsize= (12,8))
plt.imshow(wordcloud, interpolation = 'bilinear',cmap='viridis')
plt.axis('off')
```

### LSTM 

```python
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.layers import TimeDistributed
```

```python
from sklearn.model_selection import train_test_split
text = df['text'].values
labels = df['sentiment'].values
text_train, text_test, y_train, y_test = train_test_split(text, labels)
print(text_train.shape, text_test.shape, y_train.shape, y_test.shape)
```

```python
vocab_size = 20000
maxlen = 500
embedding_size = 32

tokenizer = Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(text)

X_train = tokenizer.texts_to_sequences(text_train)
x_test = tokenizer.texts_to_sequences(text_test)

X_train = pad_sequences(X_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)
```

```python
model = Sequential()
model.add(Embedding(vocab_size, embedding_size, input_length=maxlen))
model.add(LSTM(128))
#model.add(Dense(64, activation='relu'))
#model.add(Dense(1, activation='softmax'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        min_delta=1e-4,
        patience=3,
        verbose=1
    ),
    ModelCheckpoint(
        filepath='weights.h5',
        monitor='val_accuracy', 
        mode='max', 
        save_best_only=True,
        save_weights_only=True,
        verbose=1
    )
]
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, batch_size=180, validation_data=(x_test, y_test), epochs=10, callbacks=callbacks)
```

```python
def classification_accuracy_report(actual,predicted):
    print ("\nAccuracy Score on Test Data: ", accuracy_score(actual,predicted))
    print ("\nConfusion Matrix on Test Data:\n", confusion_matrix(actual,predicted))
    print ("\nClassification Report on Test Data:\n",classification_report(actual,predicted))
```

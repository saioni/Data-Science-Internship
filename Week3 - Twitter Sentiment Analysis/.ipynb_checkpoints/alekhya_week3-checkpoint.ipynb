```python
#Loading all Packages
import pymysql
import pandas as pd
pd.options.display.max_rows=10
pd.options.display.max_columns=50
import numpy as np
n=40
import warnings
warnings.filterwarnings('ignore')
import time
import sklearn
from sklearn import svm
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import roc_curve,auc
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, precision_score
%matplotlib inline
```

```python
print("==================== Data Loading ======================")
start_time = time.time()
df=pd.read_csv('Cleaned_Data.csv')
df.info()
print("==================== Data Loaded In ======================")
print("Duration: --- %s Minutes ---" % ((time.time() - start_time)/60))
```

```python
df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# create the transform
vectorizer = CountVectorizer()
# tokenize and build vocab
vectorizer.fit(df['text'].values.astype('U'))
# summarize
print(vectorizer.vocabulary_)
# encode document
vector = vectorizer.transform(df['text'].values.astype('U'))
# summarize encoded vector
print(vector.shape)
print(type(vector))
print(vector.toarray())
X=vector
print(X)
```

```python
#start_time = time.time()
r_state=52
train, test = sklearn.model_selection.train_test_split(df, train_size = 0.75,random_state=r_state)
train, val = sklearn.model_selection.train_test_split(train, train_size = 0.75,random_state=r_state)
print("For Main Data Set :",df['sentiment'].count())
print("For Train Set :",train['sentiment'].count())
print("For Validation Set :",val['sentiment'].count())
print("For Test Set :",test['sentiment'].count())

x_train=train.drop(train.columns[1], axis = 1)
x_val=val.drop(val.columns[1], axis = 1)
x_test=test.drop(test.columns[1], axis = 1)
y_train=train['sentiment']
y_val=val['sentiment']
y_test=test['sentiment']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression(C=10,max_iter=750,random_state=52)
logreg.fit(x_train, y_train)
```

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=52)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, x_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
dataset = pd.read_csv('C:/Users/dell/Desktop/Internship/DS_Internship_Task_Week-1/DS_Internship_Task_Week-1_/EDA-OnlineNewsPopularity/OnlineNewsPopularity.csv')
dataset.head() #print first 5 rows
```

```python
#check no of rows and columns (rows,columns)
dataset.shape
```

```python
dataset.columns = dataset.columns.str.lstrip()
```

### This is to strip the extra space on the left side, for example converts ' shares' to 'shares' for the ease of use.

```python
#information about the dataset - no of entries , null or not, type of data
dataset.info()
```

```python
#statistics of target attribute (shares)
stat_shares = dataset[dataset.keys()[-1]]
stat_shares.describe()
```

### Here, clear difference can be seen between minimum and maximum value which tells us that their will be outliers present.

```python
#statistics of all the attributes
dataset.describe()
```

```python
dataset=dataset.drop(['url'],axis=1)
# The url column isn't a important feature in analysing the popularity of the news, so can be dropped.
```

```python
#Detecting Outliers/anomaly
#The dots outside the boundary are outliers
for i in dataset.columns:
    sb.boxplot(dataset[i])
    plt.show()
```

### Box plots indicates how much the values in the data are spread out, gives information about minimum, first quartile, median, third quartile and maximum. This indicates the limits beyond which all the points are treated as an anomaly/outliers

```python
#Removing Outliers
#Inter Quartile method for removing outliers 
Q1 = dataset.quantile(q=0.25) 
Q3 = dataset.quantile(q=0.75)
IQR = Q3-Q1
IQR
```

```python
sorted_shares = dataset.sort_values('shares') 
median = sorted_shares['shares'].median() 
q1 = sorted_shares['shares'].quantile(q=0.25) 
q3 = sorted_shares['shares'].quantile(q=0.75) 
iqr = q3-q1
Inner_bound1 = q1-(iqr*1.5) 
print(f'Inner Boundary 1 = {Inner_bound1}')
Inner_bound2 = q3+(iqr*1.5)  
print(f'Inner Boundary 2 = {Inner_bound2}')
Outer_bound1 = q1-(iqr*3)    
print(f'Outer Boundary 1 = {Outer_bound1}')
Outer_bound2 = q3+(iqr*3)   
print(f'Outer Boundary 2 = {Outer_bound2}')
```

```python
new_data = dataset[dataset['shares']<=Outer_bound2]
print(f'Data before Removing Outliers = {dataset.shape}')
print(f'Data after Removing Outliers = {new_data.shape}')
print(f'Number of Outliers = {dataset.shape[0] - new_data.shape[0]}')
```

```python
#check distribution of no of shares, use color attribute to change colours
sb.distplot(new_data['shares'],color='g')
```

### The above distribution plot is rightly skewed which means  mean>mode, because few high shares tend to shift the mean value to right. Much of the higher shares are clustered towards the left.

```python
# correlation map
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 50,50
sb.heatmap(dataset.corr(),annot=True,xticklabels=dataset.columns,
        yticklabels=dataset.columns)
sb.set(font_scale=2)
plt.tick_params(axis='x', labelsize=27)
plt.tick_params(axis='y', labelsize=27)
```

### Correlation map draws a matrix with all the attributes and clearly depicts the relation between them. It ignores any non-numeric attribute. The box in orange shows high values which shows those two attributes are strongly positively related.

```python
columns_channels=new_data.columns.values[17:25]
new_data[columns_channels].sum().plot(kind='pie', autopct='%1.1f%%')
sb.set(rc={'figure.figsize':(10, 8)});
```

### Most of the news published were on tuesday, wednesday and thursday.

```python
# popularity based on publishing day
columns_day = new_data.columns.values[17:25]
unpop=new_data[new_data['shares']<1400]
pop=new_data[new_data['shares']>=1400]
unpop_day = unpop[columns_day].sum().values
pop_day = pop[columns_day].sum().values

import matplotlib.pyplot as pl
from IPython import get_ipython
get_ipython().run_line_magic('matplotlib', 'inline')

fig = pl.figure(figsize = (13,5))
pl.title("Popularity of news over different day of week", fontsize = 16)
pl.bar(np.arange(len(columns_day)), pop_day, width = 0.3, align="center", color = 'b', \
          label = "popular")
pl.bar(np.arange(len(columns_day)) - 0.3, unpop_day, width = 0.3, align = "center", color = 'g', \
          label = "unpopular")
pl.xticks(np.arange(len(columns_day)), columns_day)
pl.ylabel("Count", fontsize = 14)
pl.xlabel("Days of week", fontsize = 14)
    
pl.legend(loc = 'upper right')
pl.tight_layout()
#pl.savefig("days.pdf")
pl.show()
```

### From the above bar chart, analysis can be made that news published on tuesday,wednesday,thursday and on the weekends are much popular compared to other days of the week.
